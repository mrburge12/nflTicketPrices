```python
#Load the json data
with open('C:/Users/mrbur/Desktop/nflTicketPrices/data/jsonFiles/sg_2018-09-05.json', errors = 'ignore') as json_data:
    data = json.load(json_data)
    data = str(data) #convert data to string in order to clean and manipulate later
data
```

```python
#stringClean function takes in string from json file, cleans, and returns a list of values
patternList = ['\s','\"', '"', '\[\"','\{','\}','\]','\'']
patternList

def stringClean(s): 
    for pattern in patternList:
        s = re.sub(pattern,'',s) #Remove all unwanted characters from a string
    s = s.split(' n ') #take out the n and spaces and split the words into elements of list
    s = [x.strip(' ') for x in s] #take out leading and trailing spaces
    s = list(filter(None,s)) #filter out empty elements of list
    return s

#main_dict = {}
#for string_to_clean in list_of_strings_to_clean
#apply function to 
cleanDataList = stringClean(data)
x = str(cleanDataList).split(",")
y = [x1.split(":") for x1 in x.split("\n")]
_dict = {}
for y1 in y:
    _dict[y1[0]]=y1[1]
    #main_dict[string_index] = _dict
    
    
```
